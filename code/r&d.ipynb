```python
def adf_test(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])
    print('Is the time series stationary?', 'No (reject null hypothesis)' if result[1] < 0.05 else 'Yes (fail to reject null hypothesis)')
```

```python

# Perform ADF test on the original 'active power' time series
print("ADF Test for Original 'Active Power' Time Series:")
# Set the fraction you want to keep (e.g., 0.5 for 50%)


# Use the sample method to randomly select a fraction of rows
fraction_to_keep = 0.1


# Use the sample method to randomly select a fraction of rows
fractional_df = df_fridge.sample(frac=fraction_to_keep, random_state=42)
adf_test(fractional_df['active powerr'])


# Plot rolling mean and rolling standard deviation
plt.figure(figsize=(12, 6))
plt.plot(df_fridge.index, df_fridge['active powerr'], label='Original Data - Active Power')
plt.plot(df_fridge['active powerr'].rolling(window=12).mean(), label='Rolling Mean (window=12)')
plt.plot(df_fridge['active powerr'].rolling(window=12).std(), label='Rolling Std (window=12)')
plt.title('Original Time Series Data - Active Power with Rolling Mean and Rolling Std')
plt.xlabel('Timestamp')
plt.ylabel('Active Power')
plt.legend()
plt.show()
```

```python
def adf_test(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])
    print('Is the time series stationary?', 'No (reject null hypothesis)' if result[1] < 0.05 else 'Yes (fail to reject null hypothesis)')
# Perform ADF test on the original 'active power' time series
print("ADF Test for Original 'Active Power' Time Series:")
# Set the fraction you want to keep (e.g., 0.5 for 50%)
fraction_to_keep = 0.1

# Use the sample method to randomly select a fraction of rows
fractional_df = df.sample(frac=fraction_to_keep, random_state=42)
adf_test(fractional_df['active power '])

# Plot rolling mean and rolling standard deviation
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['active power '], label='Original Data - Active Power')
plt.plot(df['active power '].rolling(window=12).mean(), label='Rolling Mean (window=12)')
plt.plot(df['active power '].rolling(window=12).std(), label='Rolling Std (window=12)')
plt.title('Original Time Series Data - Active Power with Rolling Mean and Rolling Std')
plt.xlabel('Timestamp')
plt.ylabel('Active Power')
plt.legend()
plt.show()
```
